```
#Imports

import sklearn 
import numpy as np
import pandas as pd
import random 
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```
# creating data

data={
'size':[], 
'color':[], 
'gender':[], 
'price':[], 
'weight':[], 
'bought':[] 
} 
for _ in range(20):
  data['size'].append(random.choice(["XS","S","M","L","XL"]))
  data['color'].append(random.choice(["BLACK","WHITE","GREEN","YELLOW","BLUE","RED"]))
  data['gender'].append(random.choice(["MALE","FEMALE","DIVERSE"]))
  data['price'].append(round(random.uniform(5,2000),2))
  data['weight'].append(random.randint(40,200))
  data['bought'].append(random.choice(["YES","NO"]))

df_raw=pd.DataFrame(data=data) 
df=df_raw.copy() 

for col in ['size','color','gender','bought']: 
  df[col]=df[col].astype('category') 

for col in ['weight']: 
  df[col]=df[col].astype('float') 

df.info()
df.describe(include=['float']).T 
```

```
# Transforming String type data

le=LabelEncoder()
le.fit_transform(df['bought'])
print(le.classes_)

df['bought']=le.fit_transform(df['bought'])
print(df)
```

```
#categorizing input data
pd.get_dummies(data=df, drop_first = True) 
```

```
# standardizing digits
scaler = StandardScaler() 
df[['price', 'weight']] = scaler.fit_transform(df[['price', 'weight']]) 

print(df)
print(df_raw)
```
