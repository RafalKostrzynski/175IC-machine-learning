```
# Load data 
from sklearn.datasets import load_iris
import numpy as np

np.random.seed(50)
raw_data = load_iris()
all_data = raw_data.copy()

data=all_data['data']
target = all_data['target']
feature_names = all_data['feature_names']
target_names = all_data['target_names']
```

```
# change data type

data = np.array(data[:,:2])
target = np.array(list(map(int,target)))
```

```
# splitt in train and test data
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(data,target)
```

```
# create plot
from sklearn.svm import SVC
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt
from random import randint

def plot(C, kernel):
    plt.figure(figsize=(8,6))
    svc = SVC(C=C, kernel=kernel)
    svc.fit(X_train, Y_train)
    plot_decision_regions(X_train, Y_train, svc)

    plt.xlabel('sepal length (cm)')
    plt.ylabel('sepal width (cm)')
    plt.text(x=3.5,y=5,s=f'C={C}\nKernel={kernel}')
    plt.title(f"SVC train accuracy: {svc.score(X_test,Y_test):.4f}")
    plt.show()


kernels=['linear', 'poly', 'rbf']
for kernel in kernels:
  plot(randint(1,3),kernel)
  plot(randint(4,6),kernel)
  plot(randint(7,9),kernel)



```
